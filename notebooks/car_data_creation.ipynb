```python
import pandas as pd
```

## 1 - Reading the main GreeMko Data

```python
df = pd.read_csv("data.txt",sep=';',encoding="ISO-8859-1",quotechar=None, quoting=3)
```

```python
df
```

```python
df['SourceDescription'].value_counts()
```

```python
distinct_orga = df.groupby(by=['OrganizationalUnitID']).count()
```

## 2 - Get the distinct OrganizationalUnitID (distinct locations)

```python
distinct_orga = distinct_orga.reset_index()
distinct_orga = distinct_orga[['OrganizationalUnitID']]
distinct_orga
```

```python
distinct_orga = distinct_orga.astype({"OrganizationalUnitID": object})
```

```python
#check the number of distinct locations

distinct_orga['OrganizationalUnitID'].nunique()
```

## 3 - Apply random number for each distinct OrganizationalUnitID

```python
#for each rows random number from 1 to 30

import random

distinct_orga['number_cars'] = [ random.randint(0,20)  for k in distinct_orga['OrganizationalUnitID']]

distinct_orga
```

## 4 - Reading the file containing car information

```python
vehicle_df = pd.read_csv("vehicle_info.csv")
```

```python
#each distinct location has a unique number
vehicle_df.columns
```

```python
vehicle_df['Ewltp (g/km)'].isna().sum()
```

```python
vehicle_df = vehicle_df.loc[vehicle_df['Ewltp (g/km)'].isna() == False]
```

```python
vehicle_df['Ewltp (g/km)'].isna().sum()
```

```python
vehicle_df
```

```python
vehicle_df_short = vehicle_df[['ID','Mk','Cn','Ewltp (g/km)','Ft']]
vehicle_df_short 
```

```python
vehicle_df_short['ID'].min()
```

```python
vehicle_df_short['ID'].max()
```

## 5 - Assign random cars for each organization

```python
distinct_orga
```

```python
cars_list = vehicle_df_short['ID'].tolist()
final_list = []
for index, row in distinct_orga.iterrows():
    vehicle_list = []
    for elem in range(row['number_cars']):
        vehicle_list.append(random.choice(cars_list))
    final_list.append(vehicle_list)
        
    
```

```python
final_list_series = pd.Series(final_list)
```

```python
distinct_orga['vehicle_list'] = final_list_series
```

```python
distinct_orga
```

## 6 - Transpose the cars to new rows

```python
distinct_orga_vehicles = distinct_orga.explode('vehicle_list').reset_index(drop=True)
distinct_orga_vehicles
```

```python
distinct_orga_vehicles = pd.merge(distinct_orga_vehicles, vehicle_df_short, left_on='vehicle_list', right_on='ID', how='inner')
```

```python
distinct_orga_vehicles.drop(columns='vehicle_list', inplace=True)
```

```python
distinct_orga_vehicles.rename(columns={'ID':'vehicle_id',
                                       'Mk':'maker',
                                       'Cn':'commercial_name',
                                       'Ewltp (g/km)':'CO2(g/km)',
                                       'Ft':'fuel_type'}, inplace=True)
```

```python
distinct_orga_vehicles
```

## 7- Group by and merging

```python
orga_year_month = df.groupby(by=['OrganizationalUnitID','Year','Month']).count()
```

```python
orga_year_month = orga_year_month.reset_index()

orga_year_month = orga_year_month[['OrganizationalUnitID','Year','Month']]
```

```python
orga_year_month = orga_year_month.astype({"OrganizationalUnitID": object,
                                          "Year": object, 
                                          "Month": object})
```

```python
orga_year_month 
```

```python
vehicle_data = pd.merge(distinct_orga_vehicles,orga_year_month, on='OrganizationalUnitID', how='inner')
```

```python
vehicle_data
```

## 8- Assign KM and calculate the CO2

```python
from scipy.stats import truncnorm

def get_truncated_normal(mean=0, sd=1, low=0, upp=10):
    return truncnorm(
        (low - mean) / sd, (upp - mean) / sd, loc=mean, scale=sd)
```

```python
X = get_truncated_normal(mean=2000, sd=1000, low=200, upp=4000)
```

```python
random_km = X.rvs(len(vehicle_data))
```

```python
vehicle_data['km'] = pd.Series(random_km).round(2)
```

```python
vehicle_data
```

```python
vehicle_data['CO2_emissions_kg'] = round(vehicle_data['km']*vehicle_data['CO2(g/km)']/1000)
```

```python
vehicle_data
```
